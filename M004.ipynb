### Weitere Fehler im Datenset

--> Datentypen

```python
data.info()
```

```python
data["Fert.Rate"] == "N.A."
```

```python
data[data["Fert.Rate"] == "N.A."]
```

```python
data[data["Fert.Rate"] == "N.A."]["Fert.Rate"]
```

```python
data.loc[data["Fert.Rate"] == "N.A.", "Fert.Rate"] = 0  # Alle Spalten die N.A. enthalten mit 0 ersetzen, WENN die Spalte "Fert.Rate" ist
```

### Eigene Funktion um N.A. zu beheben

```python
def fixNA(columnName, newValue):
    data.loc[data[columnName] == "N.A.", columnName] = newValue
```

```python
fixNA("Med.Age", 0)
```

```python
data
```

```python
data.columns  # Alle Spaltennamen nehmen
```

```python
# Alle N.A. Werte aus allen Spalten durch 0 ersetzen
for colName in data.columns:
    fixNA(colName, 0)
```

```python
data
```

#### Prozentzeichen

Alle Prozente in einer gegebenen Spalte ersetzen

```python
data["YearlyChange"].replace(" %", "", regex=True)  # regex=True damit der Code funktioniert
```

```python
data
```

```python
data["YearlyChange"] = data["YearlyChange"].replace(" %", "", regex=True)
```

```python
data
```

```python
def fixPct(columnName):
    data[columnName] = data[columnName].replace(" %", "", regex=True)
```

```python
for colName in data.columns:
    fixPct(colName)
```

```python
data
```

#### Datentypen fixen

Die Datentypen in dem Datenset sind weiterhin strings -> keine Berechnungen möglich

```python
data.info()
```

convert_dtypes: Automatisch Typen konvertieren

Funktioniert nur mäßig

```python
data.convert_dtypes().info()
```

astype(Typ): Wandelt die gegebene Spalte in den entsprechenden Typen um

```python
data["YearlyChange"].astype(float)
```

```python
data["YearlyChange"] = data["YearlyChange"].astype(float)
```

```python
data.info()
```

```python
def fixType(columnName, targetType):
    data[columnName] = data[columnName].astype(targetType)
```

```python
fixType("Fert.Rate", float)
fixType("Med.Age", int)
fixType("UrbanPop %", float)
fixType("WorldShare", float)
```
