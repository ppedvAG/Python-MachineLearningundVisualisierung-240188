### Zwei verschiedene Möglichkeiten um Diagramme darzustellen

- Globaler Modus
    - plt.\<Funktion>
    - plt.plot(...)
 
- Objektorientierte Modus
    - a, b = plt.subplots(...)
    - a.\<Funktion>
    - b.\<Funktion>

```python
plt.title("Linien")
plt.plot([1, 2, 3])
plt.plot([4, 5, 6], color="red")
plt.plot([7, 8, 9])
```

```python
plt.subplot(2, 2, 1)  # Zeile, Spalte, Nr
plt.plot(np.random.random(100))

plt.subplot(2, 2, 2)
plt.plot(np.random.random(100))

plt.subplot(2, 2, 3)
plt.plot(np.random.random(100))

plt.subplot(2, 2, 4)
plt.plot(np.random.random(100))
plt.title("Die Vierte Linie")
```

#### Der show Befehl

Schließt die Erstellung von einem Plot ab

```python
plt.title("Linien")
plt.plot([1, 2, 3])
plt.show()
plt.plot([4, 5, 6], color="red")
plt.show()
plt.plot([7, 8, 9])  # 3 Plots statt einem
```

#### Objektorientierter Modus

Jeder Graph kann hier direkt angesprochen werden

```python
figure, ((a, b), (c, d)) = plt.subplots(2, 2)
a.plot(np.random.random(10))

d.set_title("4")
d.plot(np.random.random(10))
```

### Beispiele

#### Bar Chart

plt.bar

plt.barh

```python
# Aufgabenstellung: Top 20 Länder in einem Balkendiagramm darstellen
```

```python
data = pd.read_csv("Data/PopulationData.csv", delimiter=";", thousands=",", decimal=".")
```

```python
data.rename(columns={ "Country (or dependency)": "Country", "Population(2020)": "Pop", "Density(P/Km²)": "Density", "Land Area(Km²)": "Area", "Migrants(net)": "Migrants", "UrbanPop %": "UrbanPop" }, inplace=True)
```

```python
data.drop(columns=["#"], inplace=True)
```

```python
data
```

```python
data.loc[0:19]
```

```python
plt.figure(figsize=(12, 4))
plt.bar(data["Country"].loc[0:19].values, data["Pop"].loc[0:19])
plt.grid(alpha=0.3)
plt.title("Population Top 20")
plt.xticks(rotation=75)
None
```

```python
sorted = data.loc[0:19].sort_values("Pop")

plt.figure(figsize=(12, 4))
plt.barh(sorted["Country"].values, sorted["Pop"], color="#FF0000")
plt.grid(alpha=0.3)
plt.title("Population Top 20")
None
```

```python
plt.figure(figsize=(12, 4))
plt.bar(np.arange(0, 20) - 0.2, data["Pop"].loc[0:19], width=0.4, label="Pop")
plt.bar(np.arange(0, 20) + 0.2, data["Pop"].loc[0:19], width=0.4, color="#00FF00", label="Pop2")
plt.grid(alpha=0.3)
plt.title("Population Top 20")
plt.xticks(rotation=75, ticks=np.arange(0, 20), labels=data["Country"].loc[0:19].values)
plt.legend()
None
```

#### Pie Chart

Teile der Gesamtheit darstellen

plt.pie

```python
def fixNA(columnName, newValue):
    data.loc[data[columnName] == "N.A.", columnName] = newValue
```

```python
def fixPct(columnName):
    data[columnName] = data[columnName].replace(" %", "", regex=True)
```

```python
def fixType(columnName, targetType):
    data[columnName] = data[columnName].astype(targetType)
```

```python
fixNA("WorldShare", 0)
```

```python
fixPct("WorldShare")
```

```python
fixType("WorldShare", np.float16)
```

```python
# Aufgabenstellung: Die Top 10 Länder in einem Tortendiagramm darstellen
```

```python
plt.figure(figsize=(7, 7))
plt.pie(data["WorldShare"].loc[0:9],
        labels=data["Country"].loc[0:9],
        rotatelabels=True,
        labeldistance=0.5,
        frame=True,
        wedgeprops={'linewidth': 2, "edgecolor": "white"})
None
```

#### Scatterplot (Streudiagramm)

Zeichnet Datenpunkte mit Kreisen unterschiedlicher Größe im Diagramm ein

plt.scatter

```python
x = np.array([x + np.random.rand() * 30 for x in range(100)])
y = np.array([x + np.random.rand() * 30 for x in range(100)])
```

```python
fig = plt.figure(figsize=(12, 4))
plt.scatter(y, x, c=(x * y), s=(x * y / 10))
plt.grid(alpha=0.5, linestyle="dotted", linewidth=1)
```
